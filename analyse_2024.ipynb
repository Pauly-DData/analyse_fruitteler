```python
import pandas as pd
import openpyxl
```

```python
df = pd.read_excel('2024_verbruik.xlsx', skiprows=[0])
df
```

```python
nan_df = df.isna()
df_clean = df.fillna(0)
df_clean

#factor to multiply the kVARH columns with is 0.8 so:
columns_to_multiply = ['Blindverbruik Consumptie (kVARh)', 'Blindverbruik Consumptie (kVARh).1','Blindverbruik Consumptie (kVARh).2']
factor = 0.8

kwh_df = df_clean

for column in columns_to_multiply:
    kwh_df[column] = kwh_df[column] * factor

kwh_df
    
```

```python
#Optellen van het vebruik zodat het duidelijk is of de ramingen van ray eenigzins overeenkomen met de aangeleverde data.
Total_consumptie1 = kwh_df['Elektriciteit Consumptie (kWh)'].sum()
Total_consumptie2 = kwh_df['Elektriciteit Consumptie (kWh).1'].sum()
Total_consumptie3 = kwh_df['Elektriciteit Consumptie (kWh).2'].sum()
Total_kVARH1 = kwh_df['Blindverbruik Consumptie (kVARh)'].sum()
Total_kVARH2 = kwh_df['Blindverbruik Consumptie (kVARh).1'].sum()
Total_kVARH3 = kwh_df['Blindverbruik Consumptie (kVARh).2'].sum()

total_consumptie = Total_consumptie1 + Total_consumptie2 + Total_consumptie3
total_kVARH = Total_kVARH1 + Total_kVARH2 + Total_kVARH3

print(f'Total consumptie: {total_consumptie}')
print(f'Total kVARH: {total_kVARH}')

totaal = total_consumptie + total_kVARH
print(totaal)
```

```python
# Convert the datetime column
kwh_df['Datum-tijd tot'] = pd.to_datetime(kwh_df['Datum-tijd tot'])

# Add a new column with the rounded times to the nearest hour
kwh_df['Datum-tijd uur'] = kwh_df['Datum-tijd tot'].dt.floor('H')

# Aggregate the data by the new hour column
kwh_df_hourly = kwh_df.groupby('Datum-tijd uur').agg({
    'Elektriciteit Consumptie (kWh)': 'sum',
    'Blindverbruik Consumptie (kVARh)': 'sum',
    'Elektriciteit Consumptie (kWh).1': 'sum',
    'Blindverbruik Consumptie (kVARh).1': 'sum',
    'Elektriciteit Consumptie (kWh).2': 'sum',
    'Blindverbruik Consumptie (kVARh).2': 'sum'
}).reset_index()

# Rename the aggregated column for clarity
kwh_df_hourly.rename(columns={'Datum-tijd uur': 'StartDate'}, inplace=True)

# Show the resulting DataFrame
kwh_df_hourly
```

```python
# Bereken de som van de specifieke kolommen
sum_columns = kwh_df_hourly[['Elektriciteit Consumptie (kWh)', 'Blindverbruik Consumptie (kVARh)', 
                             'Elektriciteit Consumptie (kWh).1', 'Blindverbruik Consumptie (kVARh).1', 
                             'Elektriciteit Consumptie (kWh).2', 'Blindverbruik Consumptie (kVARh).2']].sum()

# Bereken de totale som van alle kolommen
total_sum = sum_columns.sum()

# Toon de som van de kolommen en de totale som
print("Som van elke kolom:")
print(sum_columns)
print("\nTotale som van alle kolommen bij elkaar:")
print(total_sum)
```

```python
mp = pd.read_csv('marktprijzen_2024.csv')

#mp["StartDate"] = mp["StartDate"].astype("datetime64[ns]")

mp
```

```python
# nu gaan we door naar de grote df join show

full_df = kwh_df_hourly.join(mp, how='inner', lsuffix='kwh_df_hourly', rsuffix='mp')

full_df = full_df.drop('StartDatemp', axis=1)


#full_df.to_excel("consumptie_per_uur_marktprijs_2023.xlsx")

full_df
```

```python
# Select and multiply the relevant columns by the 'marktprijs' column
price_df = full_df[['Elektriciteit Consumptie (kWh)', 'Blindverbruik Consumptie (kVARh)', 'Elektriciteit Consumptie (kWh).1', 'Blindverbruik Consumptie (kVARh).1', 'Elektriciteit Consumptie (kWh).2', 'Blindverbruik Consumptie (kVARh).2']].multiply(full_df["marktprijs"], axis="index")

# Add the 'StartDatekwh_df_hourly' column to the resulting DataFrame
price_df['StartDatekwh_df_hourly'] = full_df['StartDatekwh_df_hourly']

# Move the 'StartDatekwh_df_hourly' column to the first position
first_column = price_df.pop('StartDatekwh_df_hourly')
price_df.insert(0, 'StartDatekwh_df_hourly', first_column)

# Show the resulting DataFrame
price_df
```
