#### Deel twee van de fruittelere analyse. 

In dit deel zullen we kijken naar hoeveel de tuinder zelf heeft geproduceert zodat we dat kunnen beter kunnen kwantificeren adhv dynamische prijzen.

```python
import pandas as pd
import openpyxl
```

```python
df = pd.read_excel('Productie_kwartierdata_2023.xlsx', skiprows=[0])
df
```

```python
nan_df = df.isna()
df_clean = df.fillna(0)
df_clean

#factor to multiply the kVARH columns with is 0.8 so:
columns_to_multiply = ['Blindverbruik Productie (kVARh)', 'Blindverbruik Productie (kVARh).1','Blindverbruik Productie (kVARh).2']
factor = 0.8

kwh_df = df_clean

for column in columns_to_multiply:
    kwh_df[column] = kwh_df[column] * factor

kwh_df
```

```python
#Optellen van het vebruik zodat het duidelijk is of de ramingen van ray eenigzins overeenkomen met de aangeleverde data.

Total_consumptie1 = kwh_df['Elektriciteit Productie (kWh)'].sum()
Total_consumptie2 = kwh_df['Elektriciteit Productie (kWh).1'].sum()
Total_consumptie3 = kwh_df['Elektriciteit Productie (kWh).2'].sum()

Total_kVARH1 =  kwh_df['Blindverbruik Productie (kVARh)'].sum()
Total_kVARH2 =  kwh_df['Blindverbruik Productie (kVARh).1'].sum()
Total_kVARH3 =  kwh_df['Blindverbruik Productie (kVARh).2'].sum()

print(Total_consumptie1)
print(Total_consumptie2)
print(Total_consumptie3)
print(Total_kVARH1)
print(Total_kVARH2)
print(Total_kVARH3)

Totaal_kwh = (Total_consumptie1 + Total_consumptie2 + Total_consumptie3 + Total_kVARH1 + Total_kVARH2 + Total_kVARH3)
print(f'Totaal aan kwh: {Totaal_kwh}')
```

```python
kwh_df['Datum-tijd tot'] = pd.to_datetime(kwh_df['Datum-tijd tot'], format='%d-%m-%Y %H:%M')

# Voeg een nieuwe kolom toe met de afgeronde tijden naar het dichtstbijzijnde uur
kwh_df['Datum-tijd uur'] = kwh_df['Datum-tijd tot'].dt.ceil('H')

kwh_df_hourly = kwh_df.groupby('Datum-tijd uur').agg({
    
    'Elektriciteit Productie (kWh)': 'sum',
    'Blindverbruik Productie (kVARh)': 'sum',
    'Elektriciteit Productie (kWh).1': 'sum',
    'Blindverbruik Productie (kVARh).1': 'sum',
    'Elektriciteit Productie (kWh).2': 'sum',
    'Blindverbruik Productie (kVARh).2': 'sum'

    }).reset_index()


kwh_df_hourly.rename(columns={'Datum-tijd uur' : 'StartDate'}, inplace=True)

# Toon het resulterende DataFrame
kwh_df_hourly
```

```python
mp = pd.read_csv('marktprijzen_2023.csv')

mp["StartDate"] = mp["StartDate"].astype("datetime64[ns]")
```

```python
# nu gaan we door naar de grote df join show

full_df = kwh_df_hourly.join(mp, how='inner', lsuffix='kwh_df_hourly', rsuffix='mp')

full_df = full_df.drop('StartDatemp', axis=1)

full_df
```

```python
# dan gaan we nu door naar het berkenene van de prijzen van de consumptie


price_df = full_df[['Elektriciteit Productie (kWh)', 'Blindverbruik Productie (kVARh)', 'Elektriciteit Productie (kWh).1', 'Blindverbruik Productie (kVARh).1', 'Elektriciteit Productie (kWh).2', 'Blindverbruik Productie (kVARh).2']].multiply(full_df["Price_kWh"], axis="index")
price_df['StartDatekwh_df_hourly'] = full_df['StartDatekwh_df_hourly']

first_column = price_df.pop('StartDatekwh_df_hourly')
price_df.insert(0, 'StartDatekwh_df_hourly', first_column)


price_df
```

```python
kosten_onbalans = Totaal_kwh *  -0.010260
kosten_onbalans
```

```python
#sommeren van de columns om tataal te krijgen. 
Total_consumptie1 = price_df['Elektriciteit Productie (kWh)'].sum()
Total_consumptie2 = price_df['Elektriciteit Productie (kWh).1'].sum()
Total_consumptie3 = price_df['Elektriciteit Productie (kWh).2'].sum()

Total_kVARH1 =  price_df['Blindverbruik Productie (kVARh)'].sum()
Total_kVARH2 =  price_df['Blindverbruik Productie (kVARh).1'].sum()
Total_kVARH3 =  price_df['Blindverbruik Productie (kVARh).2'].sum()

#print(Total_consumptie1)
#print(Total_consumptie2)
#print(Total_consumptie3)
#print(Total_kVARH1)
#print(Total_kVARH2)
#print(Total_kVARH3)

prijs_totaal = (Total_consumptie1 + Total_consumptie2 + Total_consumptie3 + Total_kVARH1 + Total_kVARH2 + Total_kVARH3)

prijs_round = prijs_totaal.round(2)
opslag = kosten_onbalans.round(2)

oplevering = prijs_round + opslag

print(f'In het jaar 2023 met een marktprijscontract zou er voor: â‚¬ {oplevering} aan teruglevering via Prodcutie zijn geweest')
```
